```python
import matplotlib.pyplot as plt

# Data: Store types and their respective counts
store_types = ['Restaurant/Cafes/Bars', 'Grocery Stores', 'Clothing Stores', 
               'Department Stores', 'Electronics', 
               'Pharmacies/Health Stores', 'Other']
store_counts = [67, 25, 6, 7, 3, 4, 9]

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Set the size of the figure
bars = plt.bar(store_types, store_counts, color='skyblue', edgecolor='black', linewidth=1.5)  # Add black border

# Adding title and labels
plt.title('Distribution of Store Types', fontsize=16)
plt.ylabel('Number of Stores', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set the y-axis limit to add space between the bars and the top frame
plt.ylim(0, max(store_counts) + 10)  # Increase the upper limit for more space

# Add numbers on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, str(height), ha='center', va='bottom', fontsize=12)

# Adjust layout to fit everything
plt.tight_layout()

# Save the figure as a PNG file
plt.savefig('store_types_barchart.png', dpi=300)  # Save with high resolution

# Show the plot
plt.show()
```
